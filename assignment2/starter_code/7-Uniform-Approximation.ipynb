# 4

To solve the XOR case, one solution is (from the Deep Learning book, http://www.deeplearningbook.org/) is with one 2-neuron hidden layer

$$f(x;W, c, w, b) = w^T \max\{0, W^T x + c\} + b$$

$\max$ here is a ReLU element-wise nonlinear transformation. 

\begin{align*}
W &= \begin{bmatrix}
 1 & 1\\ 
 1 & 1
\end{bmatrix} \\
c &= \begin{bmatrix}
 0\\ 
 -1
\end{bmatrix} \\
w &= \begin{bmatrix}
 1 \\ 
 -2
\end{bmatrix} \\
b &= 0
\end{align*}

Considering four input cases simultaneously, let

\begin{align*}
X &= \begin{bmatrix}
 0 & 0 & 1 & 1\\ 
 0 & 1 & 0 & 1\\ 
\end{bmatrix}
\end{align*}

then

\begin{align*}
W^TX &= \begin{bmatrix}
 0 & 1 & 1 & 2\\ 
 0 & 1 & 1 & 2\\ 
\end{bmatrix} \\
W^TX \hat{+} c &= \begin{bmatrix}
 0 & 1 & 1 & 2\\ 
 -1 & 0 & 0 & 1\\ 
\end{bmatrix} \\
\max\{W^TX \hat{+} c\} &= \begin{bmatrix}
 0 & 1 & 1 & 2\\ 
 0 & 0 & 0 & 1\\ 
\end{bmatrix} \\
w^T \max\{W^TX \hat{+} c\} &= \begin{bmatrix}
 0 & 1 & 1 & 0\\ 
\end{bmatrix}
\end{align*}

$\hat{+}$ means it's not a normal addition since $c$ is broadcasted to all columns.

So XOR becomes linearly separable after transformation by the hidden layer (i.e. up to the third equation above).

# 5

The reason to considering multiple hidden layers could be for practical purpose because a very wide layer may not be easily trainable. For example, when the structure in the data could be effectively represented in a hierarchical manner, modeling such a structure with a flattened one layer could be more difficult than using multiple layers.
